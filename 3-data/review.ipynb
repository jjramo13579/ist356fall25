```python
#Q1

import pandas as pd
exams = pd.read_csv('https://raw.githubusercontent.com/mafudge/datasets/refs/heads/master/exam-scores/exam-scores.csv')
exams.sample(5)
```

```python
# Use this syntax are using the same column for grouping and aggregation
# Class_Section' is both an index level and a column label
exams_by = exams.groupby(by=['Class_Section']).agg({ 'Class_Section': 'count', 'Student_Score': 'mean' })
exams_by
```

```python
import pandas as pd
data = {'Region': ['East', 'West', 'East', 'West', 'East', 'East'],
        'Product': ['A', 'B', 'A', 'A', 'B', 'B'],
        'Quantity': [10, 5, 12, 8, 6, 9],
        'Price': [2.0, 4.0, 2.0, 4.0, 3.0, 3.0]}
df_sales = pd.DataFrame(data)
df_sales
```

```python
# Q2

'''
Option D correctly groups by Region and uses the agg() function with a dictionary to apply sum to the 
Quantity column and mean to the Price column.

'''

df_sales.groupby('Region').agg({'Quantity': 'sum', 'Price': 'mean'})


```

```python
# Q3

'''
C uses pivot_table with the correct index (Region), columns (Product), values (Quantity), and the aggregation function (aggfunc= sum) 

'''

df_sales.pivot_table(index='Region', columns='Product', values='Quantity', aggfunc='sum') 



```

```python
data_wide = {'City': ['London', 'Paris', 'Tokyo'],
             'Q1_Sales': [100, 150, 200],
             'Q2_Sales': [110, 160, 210],
             'Q3_Sales': [120, 170, 220]}
df_wide = pd.DataFrame(data_wide)
df_wide 
```

```python
# Q4

'''
melt()is used to go from wide to long format. 
id_vars='City'keeps 'City' as the identifying column. The values in this columns are repeated to match all the new rows.
var_name='Quarter' names the new column that holds the old column headers ('Q1_Sales', etc.).
value_name='Sales'names the new column that holds the data values (100, 110, 120, etc.).

'''

df_wide.melt(id_vars='City', var_name='Quarter', value_name='Sales') 



```

```python
import pandas as pd
data = {'Department': ['HR', 'IT', 'HR', 'IT', 'Sales', 'Sales', 'HR', 'IT'],
        'EmployeeID': [101, 201, 102, 202, 301, 302, 103, 203],
        'Salary': [50000, 75000, 55000, 80000, 60000, 65000, 50000, 90000],
        'YearsService': [2, 5, 3, 6, 1, 2, 2, 7]}
df_employees = pd.DataFrame(data)

df_employees 
```

```python
# Q6

'''
Groups the DataFrame by the Department column,
Selects the Salary column for aggregation, 
and then applies the max() function to find the largest salary in each group.

'''

df_employees.groupby('Department')['Salary'].max() 




```

```python
# Q7
'''
When a single column is selected ([Salary]), 
the aggregation functions in the list are applied to it. 
In this context, size will return the number of items in the group (the employee count), 
and mean will return the average salary.

'''



df_employees.groupby('Department')['Salary'].agg(['size', 'mean']) 

```

```python
# Q8


'''
Using a dictionary in agg() is used to apply different aggregation functions to different columns. 
The key is the column name to aggregate, and the value is the function to apply.

'''

df_employees.groupby('Department').agg({'Salary': 'sum', 'YearsService': 'max'}) 
```

```python
import pandas as pd
data_a = {'CustomerID': [1, 2, 3, 4, 5],
          'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
          'Region': ['North', 'South', 'North', 'West', 'East']}
df_customers = pd.DataFrame(data_a)
df_customers 
```

```python
data_b = {'OrderID': ['X100', 'X101', 'X102', 'X103', 'X104', 'X105'],
          'Customer_ID': [1, 3, 3, 6, 2, 2],
          'Amount': [50.0, 75.0, 10.0, 100.0, 25.0, 90.0]}
df_orders = pd.DataFrame(data_b)
df_orders 
```

```python
'''
Inner join is the default behavior for pd.merge(), so specifying how='inner'is optional. 
You need to correctly identify the linking columns: CustomerID in the left table 
and Customer_ID in the right table 
using left_on and right_on.
Only rows with matching CustomerID/Customer_ID values will be included in the result.

'''


pd.merge(df_customers, df_orders, left_on='CustomerID', right_on='Customer_ID') 
```

```python
'''
A left join keeps every row from the left table (df_customers).
CustomerID 1, 2, 4, 5 appear once in df_customers.
CustomerID 3 appears once in df_customer but has two matching orders (X101, X102) in df_orders. 
This results in two rows for Charlie (ID 3) in the final merged DataFrame.

'''





result_left = pd.merge(df_customers, df_orders, left_on='CustomerID', right_on='Customer_ID', how='left')
result_left
```
